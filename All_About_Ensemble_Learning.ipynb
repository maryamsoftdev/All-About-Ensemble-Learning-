**Ensemble learning** is a machine learning technique where multiple week models a train on the data and then combine to create a single strong model.

**why**:Ensemble learning reduced bias and variance while limiting the risk of overfitting

**Bias** is the model tendency to consistently learn the wrong things by not taking into account all the information in the data

**High Bias** is a result of the algorithm missing the relevant relationship between features and target outputs

**Variance** refers to an algorithm sensitivity to small fluctuations in the training data

**High Variance** is the result of algorithms fitting to random noise in the training data

**Boosting** is an ensemble method that sequentially train number of weak models (often trees) to create one strong model.

1)The sequential nature of training is critical as boosting effectively learned from its mistake with each iteration

2)Computationally expensive to use for its sequential nature

3)weighted voting

4)Boosting address the under fitting problem of the model where high bais and low variance  (reducing the bias.)

5)weak and shallow models

**Bagging**:

1)The main task of bagging is decrease the bias but not variance.

2)Each of the model is built here independently.

3)It is trying to solve by over fitting problem.
4)low bias high varience models(reducing varience)
5)weak model but deep models
**Stacking:**

Stacking, or stacked generalization, is a model ensembling technique that combines multiple base models using a meta-learner. Instead of relying on a single method like voting or averaging, stacking learns how to combine the predictions of the base models to make final predictions. It often involves splitting the training data into multiple folds, where each fold is used to train different base models, and then using another fold to train the meta-learner on the predictions of these base models. Stacking can often result in improved performance compared to individual base models
2)Meta model for final prediction

3)same as bagging, low bias high varience models(reducing varience)
4) each model is trained on full data unlike subsets in other models

# Example is random forest model by using bagging.

```python
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier  # Import RandomForestClassifier
import numpy as np

# Load the dataset
data = datasets.load_wine(as_frame=True)
X = data.data
y = data.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=22)

# Define a range of estimator values
estimator_range = np.arange(2, 21, 2)

# Lists to store models and scores
models = []
scores = []

# Iterate over the estimator range
for n_estimators in estimator_range:
    # Create a Random Forest classifier
    clf = RandomForestClassifier(n_estimators=n_estimators, random_state=22)

    # Fit the model
    clf.fit(X_train, y_train)

    # Predict on the test set
    y_pred = clf.predict(X_test)

    # Calculate the accuracy score
    score = accuracy_score(y_true=y_test, y_pred=y_pred)

    # Output the number of estimators and the corresponding score
    print(f"Number of estimators: {n_estimators}, Score: {score}")

    # Append the model and score to their respective lists
    models.append(clf)
    scores.append(score)

# Generate the plot of the scores against the number of estimators
plt.figure(figsize=(9, 6))
plt.plot(estimator_range, scores)

# Adjust labels and font (to make them visible)
plt.xlabel("n_estimators", fontsize=18)
plt.ylabel("score", fontsize=18)
plt.tick_params(labelsize=16)

# Show the plot
plt.show()

```

# Stacking Implementation

```python
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.ensemble import StackingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
import numpy as np

# Load the dataset
data = datasets.load_wine(as_frame=True)
X = data.data
y = data.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=22)

# Define base classifiers
base_classifiers = [
    ('random_forest', RandomForestClassifier(n_estimators=10, random_state=22)),
    ('gradient_boosting', GradientBoostingClassifier(n_estimators=10, random_state=22))
]

# Define the meta-learner
meta_learner = DecisionTreeClassifier(random_state=22)

# Create the stacking classifier
clf = StackingClassifier(estimators=base_classifiers, final_estimator=meta_learner)

# Fit the model
clf.fit(X_train, y_train)

# Predict on the test set
y_pred = clf.predict(X_test)

# Calculate the accuracy score
score = accuracy_score(y_true=y_test, y_pred=y_pred)
print("Stacking Classifier Accuracy:", score)


```

```python
# Create the stacking classifier
stacking_clf = StackingClassifier(estimators=base_classifiers)

# Now you can access the parameters using get_params()
params = stacking_clf.get_params()
print(params)
```
